```
nota1=float(input("Digite a nota 1: "))
nota2=float(input("Digite a nota 2: "))
nota3=float(input("Digite a nota 3: "))
nota4=float(input("Digite a nota 4: "))
print(f"A média final é {(nota1+nota2+nota3+nota4)/4:.2f}")
```

Crie um algoritmo em Python que peça 2 números inteiros e um número real. Calcule e mostre:
* a soma do dobro do primeiro com metade do segundo .
* a soma do triplo do primeiro com o terceiro.
* o terceiro elevado ao cubo.

```
x=int(input("Digite um número inteiro: "))
y=int(input("Digite um número inteiro: "))
z=float(input("Digite um número real: "))
print(f"A soma do dobro do primeiro com a metade do segundo é {2*x+y/2:.2f}")
print(f"A soma do triplo do primeiro com o terceiro vale {3*x+z:.2f}")
print(f"o terceiro elevado ao cubo é {z**3:.2f}")
```
